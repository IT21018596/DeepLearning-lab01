```python
import pandas as pd
df = pd.read_csv('Au_nanoparticle_dataset.csv')
df.head(10)
```

```python
# Display the first few rows of the dataframe
print("Original DataFrame:")
print(df.head())

# List of columns to exclude
exclude_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new dataframe by filtering out the specified columns
filtered_df = df.drop(columns=exclude_columns)

# Display the first few rows of the filtered dataframe
print("\nFiltered DataFrame:")
print(filtered_df.head())
```

```python
print(filtered_df.head(20))
```

```python
include_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new dataframe by selecting the specified columns
selected_df = df[include_columns]

# Calculate statistics
mean_values = selected_df.mean()
std_values = selected_df.std()
quartile_values = selected_df.quantile([0.25, 0.5, 0.75])

# Display the results
print("Mean values:")
print(mean_values)
print("\nStandard Deviation values:")
print(std_values)
print("\nQuartile values (25th, 50th, 75th percentiles):")
print(quartile_values)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

include_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new dataframe by selecting the specified columns
selected_df = df[include_columns]

# Plot the histograms in a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot each histogram
for ax, column in zip(axes, include_columns):
    ax.hist(selected_df[column], bins=20, color='blue', alpha=0.7)
    ax.set_title(f'{column} Histogram')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

include_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new dataframe by selecting the specified columns
selected_df = df[include_columns]

# Create pairplot
sns.pairplot(selected_df)
plt.suptitle('Pairplot of N_total, N_bulk, N_surface, and R_avg', y=1.02)
plt.show()
```
