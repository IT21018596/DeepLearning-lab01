```python
# Set the scale parameter for the exponential distribution (位). For example, let's use 位 = 1.0
scale_parameter = 1.0

# Create a 100000x1 array from the exponential distribution
exp_array = np.random.exponential(scale=scale_parameter, size=(100000, 1))

# Create 100000x1 arrays from uniform and normal distributions for comparison
uniform_array = np.random.uniform(low=0.0, high=scale_parameter, size=(100000, 1))
normal_array = np.random.normal(loc=0.0, scale=scale_parameter, size=(100000, 1))
```

```python
import matplotlib.pyplot as plt
# Plot histograms
plt.figure(figsize=(15, 6))

# Histogram for exponential distribution
plt.subplot(1, 3, 1)
plt.hist(exp_array, bins=100, color='blue', alpha=0.7)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for uniform distribution
plt.subplot(1, 3, 2)
plt.hist(uniform_array, bins=100, color='green', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for normal distribution
plt.subplot(1, 3, 3)
plt.hist(normal_array, bins=100, color='red', alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid for X and Y values in the range [-5, 5]
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute Z values
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add color bar which maps values to colors
fig.colorbar(surface, shrink=0.5, aspect=5)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('$Z = X^2 + Y^2$')

plt.show()
```
